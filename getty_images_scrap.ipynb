```python
import re

def create_relative_url(base_url, i):
    relative_string = "&page="+str(i)
    next_substring = re.search(r'\b(&phrase=)\b', base_url)
    index = next_substring.start()

    l = list(base_url) 
    l.insert(index, str(relative_string)) 
    l = ''.join(l) 
    
    return l
```

```python
#base_url = "https://www.gettyimages.com.br/fotos/people-wearing-mask?phrase=people%20wearing%20mask&sort=mostpopular#license"
# https://www.gettyimages.com.br/fotos/people-wearing-mask?phrase=people%20wearing%20mask&sort=mostpopular#license

#base_url = crianca_asiatica
#base_url = crianca_negra
#base_url = crianca_branca
#base_url = adulto_negro
#base_url = adulto_branco
#base_url = asian
base_url = black_people_glasses

# +"&page="+str(i+1)+
urls = []
max_page = 8

for i in range(0, max_page):
    if i == 0:
        urls.append(base_url)
    else:
        relative_url = create_relative_url(base_url, i+1)
        urls.append(relative_url)
        
print("Number of pages: {}".format(len(urls)))
print(urls[0])
print(urls[-1])
```

```python
# Check if all urls exist
def check_url(url):
    request = requests.get(url)
    if request.status_code == 200:
        return 1
    return 0

flag = 1
for index, url in enumerate(urls):
    print("[Checking URL] {}/{}".format(index, len(urls)-1))
    if (not check_url(url)):
        print('WEB SITE INDEX {} DOES NOT EXIST'.format(index))
        flag = 0

if(flag):
    print("All urls exist!")
```

#### Get the source link of each image in urls 

```python
links = []
for index, url in enumerate(urls):
    #if index == 10: # early stop
        #break
    
    r = requests.get(url)
    soup = bs(r.content)

    mosaic = soup.find(class_="search-content__gallery-container")
    mosaic.find_all(class_="gallery-mosaic-asset__link")[-1].extract()
    images_mosaic = mosaic.find_all(class_="gallery-mosaic-asset__link")

    print("Page {}/{}, Number of images: {}".format(index+1, len(urls), len(images_mosaic)))
    for idx, link in enumerate(images_mosaic):
        links.append(link.img["src"])
```

```python
def save_in_folder(root_path, class_sub_folder, links, early_stop=-1):
    root_path = os.path.join(root_path, class_sub_folder)
    if(not os.path.isdir(root_path)):
        os.mkdir(root_path)

    for index, link in enumerate(links):
        if early_stop > 0:
            if index == early_stop: 
                break
        print("[Saving {}] {}/{}".format(class_sub_folder, index, len(links)))
        path = os.path.join(root_path, class_sub_folder+str(index)+".jpg")
        with open(path, "wb") as f:
            f.write(requests.get(link).content)
            
    print()
```
